Imports:

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

Importing and Preprocessing Data:

```python
csv_path = "Data/combined_fixed.csv"
df = pd.read_csv(csv_path)

df = df.sample(frac=1.0, random_state=17).reset_index(drop=True)

features = [
    "batch_size",
    "buffer_size",
    "learning_rate",
    "beta",
    "epsilon",
    "lambda",
    "num_epoch",
    "hidden_units",
    "num_layers",
    "gamma",
    "time_horizon",
]

target_time = "Time Elapsed/s",
target_reward = "3DBall.Environment.CumulativeReward.mean"

X = df[features]
Y_time = df[target_time]
Y_reward = df[target_reward]

X_train, X_test, Y_time_train, Y_time_test, Y_reward_train, Y_reward_test = train_test_split(X, Y_time, Y_reward, test_size=0.2, random_state=17)
```

Time prediction model:

```python
rf_time = RandomForestRegressor(n_estimators=300, random_state=17, n_jobs=-1)
rf_time.fit(X_train, Y_time_train)
```

Reward prediction model:

```python
rf_reward = RandomForestRegressor(n_estimators=300, random_state=17, n_jobs=-1)
rf_reward.fit(X_train, Y_reward_train)
```

Evaluation:

```python
rf_time_predictions = rf_time.predict(X_test)
rf_reward_predictions = rf_reward.predict(X_test)

print("Time Model MSE: ", mean_squared_error(Y_time_test, rf_time_predictions))
print("Reward Model MSE: ", mean_squared_error(Y_reward_test, rf_reward_predictions))
```
